```python
lbl_enc = LabelEncoder()
lbl_enc.fit(train4p.component_id_1_name.values)  
print lbl_enc.classes_
comp_id1_trans = lbl_enc.transform(train4p.component_id_1_name.values)
from collections import defaultdict
d = defaultdict(int)
for i in comp_id1_trans:
    d[i] += 1
print d


for i in range(len(comp_id1_trans)):
    if d[comp_id1_trans[i]] < 135:
        comp_id1_trans[i] = 1 

d1 = defaultdict(int)
for i in comp_id1_trans:
    d1[i] += 1

print d1

lbl1_map = {}
for i, lbl in enumerate(list(lbl_enc.classes_)):
    lbl1_map[lbl] = i
```

```python
lbl_enc = LabelEncoder()
lbl_enc.fit(train4p.component_id_2_name.values)  
print lbl_enc.classes_
print train4p.component_id_2_name.value_counts()
print list(lbl_enc.classes_).index('SLEEVE-ORFS')
comp_id2_trans = lbl_enc.transform(train4p.component_id_2_name.values)
from collections import defaultdict
d = defaultdict(int)
for i in comp_id2_trans:
    d[i] += 1
print d
for i in range(len(comp_id2_trans)):
    if d[comp_id2_trans[i]] < 109:
        comp_id2_trans[i] = 1 

d2 = defaultdict(int)
for i in comp_id2_trans:
    d2[i] += 1

print d2

lbl2_map = {}
for i, lbl in enumerate(list(lbl_enc.classes_)):
    lbl2_map[lbl] = i
```
